```python
import random

def get_mca_questions(context):
    if not isinstance(context, str):
        raise ValueError("Context should be a string.")

    mca_questions = []

    # Question 1
    
    q1_options = [('a', 'Stripping electrons from suitable substances'),
                  ('b', 'Reducing and removing compounds'),
                  ('c', 'Producing glucose'),
                  ('d', 'Creating NADPH and ATP')]
    q1_correct_options = ['a', 'd']
    q1 = generate_mca_question(context, "Which of the following are examples of light-dependent reactions in photosynthesis?", q1_options, q1_correct_options)
    mca_questions.append(q1)

    # Question 2
    
    q2_options = [('a', 'Oxygen'),
                  ('b', 'Hydrogen sulfide'),
                  ('c', 'Water'),
                  ('d', 'Glucose')]
    q2_correct_options = ['b', 'c']
    q2 = generate_mca_question(context, "Which of the following are sources of electrons used by the first photosynthetic organisms?", q2_options, q2_correct_options)
    mca_questions.append(q2)

    # Question 3
    
    q3_options = [('a', 'Oxygen'),
                  ('b', 'Hydrogen'),
                  ('c', 'NADPH'),
                  ('d', 'Glucose')]
    q3_correct_options = ['a', 'c']
    q3 = generate_mca_question(context, "Which of the following are byproducts of photosynthesis?", q3_options, q3_correct_options)
    mca_questions.append(q3)

    return mca_questions


def generate_mca_question(context, question, options, correct_options):
    
    # Shuffle the options
    
    random.shuffle(options)

    # Generate the question string
    
    question_str = f"{context}\n{question}\n"
    for option in options:
        question_str += f"{option[0]}. {option[1]}\n"

    # Generate the answer string
    
    answer_str = ""
    for option in options:
        if option[0] in correct_options:
            answer_str += f"{option[0]}. {option[1]}\n"

    # Combine the question and answer strings
    
    full_question = f"{question_str}\nCorrect Options: {answer_str}"

    return full_question
```

```python
context = "MCA Questions examples: More than choice per question should be an answer and is a must**"
questions = get_mca_questions(context)
for i, question in enumerate(questions):
    print(f"Question {i+1}:")
    print(question)
    print()
```
